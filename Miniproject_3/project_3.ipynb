```python
import nltk
import numpy as np
import re
from nltk.tokenize import TweetTokenizer
import matplotlib.pyplot as plt
from HMM import unsupervised_HMM
from HMM_helper import (
    text_to_wordcloud,
    states_to_wordclouds,
    parse_observations,
    sample_sentence,
    visualize_sparsities,
    animate_emission,
    obs_map_reverser)
```

```python
f = open('data/shakespeare.txt', 'r')
data = f.read()
splt = data.split("\n\n\n")
sonnets = []

for sonnet in splt:
    sonnets.append(sonnet)
```

```python
tknzr = TweetTokenizer()
sonnets = [tknzr.tokenize(sonnet)[1:] for sonnet in sonnets]
```

```python
def parse_observations_sonnets(lines):

    obs_counter = 0
    obs = []
    obs_map = {}

    for line in lines:
        obs_elem = []
        
        for word in line:
            if re.fullmatch(r'[^\w]',word):
                continue
            word = word.lower()
            if word not in obs_map:
                # Add unique words to the observations map.
                obs_map[word] = obs_counter
                obs_counter += 1
            
            # Add the encoded word.
            obs_elem.append(obs_map[word])
        
        # Add the encoded sequence.
        obs.append(obs_elem)

    return obs, obs_map
```

```python
obs, obs_map = parse_observations_sonnets(sonnets)
```

```python
sonnet2 = []
reverser = obs_map_reverser(obs_map)
for token in obs[-1]:
    sonnet2.append(' ' + reverser[token])
" ".join(sonnet2)
```
